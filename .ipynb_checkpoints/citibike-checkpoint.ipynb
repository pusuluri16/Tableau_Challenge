```python
#import the dependencies
import pandas as pd
import datetime as dt
```

```python
#verify the dataframe path
june2023 = 'data/JC-202306-citibike-tripdata.csv'
```

```python
#read the dataframe into csv
june23_df = pd.read_csv(june2023, low_memory = False)
```

```python
#check the Dtypes
june23_df.info()
```

```python
#verify the dataframe
june23_df.head()
```

```python
#drop the null values in the dataframe
june23_new_df = june23_df.dropna()
```

```python
#check the Dtypes
june23_new_df.info()
```

```python
#check the rideable_type count
june23_new_df["rideable_type"].value_counts()
```

```python
#check the member casual count
june23_new_df["member_casual"].value_counts()
```

```python
#change the rideable_type and member_casual dtype as category
june23_new_df = june23_new_df.astype({"rideable_type":"category", "member_casual":"category"})
```

```python
#verify the dtypes
june23_new_df.info()
```

```python
# change the started_at and ended-at dtypes
june23_new_df = june23_new_df.astype({"started_at":"datetime64", "ended_at":"datetime64"})
```

```python
#verify the dtypes
june23_new_df.info()
```

```python
# calculate the elapsed_time

june23_new_df["elapsed_time"] = june23_new_df["ended_at"] - june23_new_df["started_at"]
```

```python
#check the dataframe
june23_new_df.head()
```

```python
#describe the elapsed time
june23_new_df["elapsed_time"].describe()
```

```python
#make the elapsed_time to minutes

june23_new_df["elapsed_time_minutes"] = round(june23_new_df["elapsed_time"] / dt.timedelta(minutes=1))
```

```python
#change the elapsed time which is greater than or equal to one

june23_new_df = june23_new_df.loc[(june23_new_df["elapsed_time_minutes"] >= 1.0), :]
```

```python
# remove the docked_bike from the rideable_type

june23_new_df = june23_new_df.loc[(june23_new_df["rideable_type"] != "docked_bike"), :]
```

```python
#verify the dataframe
june23_new_df
```

```python
#reset the index

june23_final_df = june23_new_df.reset_index(drop=True)
```

```python
#reset the sample(95000)

june23_final_df = june23_final_df.sample(95000).reset_index(drop=True)
```

```python
#verify the dataframe
june23_final_df
```

```python
#save the dataframe to a csv file
june23_final_df.to_csv('data/citibike.csv', index= False)
```
